```python
import datetime

class Flight:
    flight_number = 0

    def __init__(self, departure_city:str, arrival_city:str, departure_time:str, seat_capacity:int=150) -> None:
        self.flight_number = Flight.flight_number
        self.departure_city = departure_city
        self.arrival_city = arrival_city
        self.departure_time = departure_time
        self.seat_capacity = seat_capacity
        self.passenger_list = []

        Flight.flight_number += 1

    def add_passenger(self, passenger:str) -> None:
        if(len(self.passenger_list) < self.seat_capacity):
            self.passenger_list.append(passenger)
        else:
            print(f"Could not add {passenger}, the flight is full")

    def remove_passenger(self, passenger:str) -> None:
        if(passenger in self.passenger_list):
            self.passenger_list.remove(passenger)
            print(f"{passenger} was removed from the flight.")
        else:
            print(f"{passenger} was not found on the flight.")

    def log_flight(self) -> None:
        data = [f"{self.flight_number}, {self.departure_city}, {self.arrival_city}\n", f"Number of passengers: {len(self.passenger_list)}\n\n"]
        with open(file = "flight_log.txt", mode = 'at', encoding='utf-8') as f:
            f.writelines(data)

    def check_availability(self) -> int:
        return self.seat_capacity - len(self.passenger_list)
    
    def __str__(self) -> str:
        return f"Flight {self.flight_number} from {self.departure_city} to {self.arrival_city}"


# TESTING ABC AIRLINE
# Create Flight
f1 = Flight("DC", "West Chester", "12:00 PM", 5)

# Add passengers
f1.add_passenger("Billy")
f1.add_passenger("Bob")
f1.add_passenger("Jones")
f1.add_passenger("Justin")

# Check availability and print flight info
print(f1.check_availability())
print(f1)

# Try to remove a passenger that doesnt exist
f1.remove_passenger("Michael")

# Log the flight
f1.log_flight()

# Add too many passengers
f1.add_passenger("Michael")
f1.add_passenger("Joe")

# Remove a passenger and add another
f1.remove_passenger("Michael")
f1.add_passenger("Joe")

# Create second flight and log it
print()
f2 = Flight("New York", "Seattle", "12:00 AM")
print(f2)
f2.log_flight()
```
