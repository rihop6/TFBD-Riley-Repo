```python
import pandas as pd
import numpy as np
path = "..\\..\\Lecture Code (my notes)\\Assets\\Data\\"
```

```python
# Read in dataframe
df = pd.read_csv(path+"imdb_top_1000.csv")

# Drop the unecessary columns
df.drop(columns=["Poster_Link","Certificate","Overview","Runtime","Meta_score","No_of_Votes"], inplace=True)
df.head()
```

```python
# Find the oldest movie
df.loc[df["Released_Year"] == df["Released_Year"].min()]
```

```python
# Find the newest movie
newest = df.loc[df["Released_Year"] == df["Released_Year"].max()]
```

```python
# Uh oh! The release year is incorrect. I think it would be prudent to take a deeper
# look and clean the data if possible.
df["Released_Year"].value_counts()
```

```python
# The values between 1920 and 2014 should be correct- we'll delete the entry with 'PG'
# From now on, we'll check the value counts to check for outliers
df.drop(newest.index, inplace=True)

# Get newest movies
df.loc[df["Released_Year"] == df["Released_Year"].max()]
```

```python
# Next up is the top 10 movies- we'll check the IMDB values first

# Check IMDB Ratings Values
df["IMDB_Rating"].value_counts()
```

```python
# Looks good, we'll get the top 10 movies now

# Get top 10 movies
df.sort_values("IMDB_Rating", ascending=False).head(10)
```

```python
# Now we need the top movie for each genre

# Check Genre values
df["Genre"].value_counts()
```

```python
# Looks good, but some of these have multiple genres. We'll want to separate these in our analysis
movies = df['Genre'].str.split(', ')

# Creates a list of each unique genre
unique_genres = []
for movie in movies:
    for genre in movie:
        if(genre not in unique_genres):
            unique_genres.append(genre)
unique_genres.sort()

# Find the best in each movie genre
for genre in unique_genres:
    # Narrow it down to movies that are in this genre
    contenders = df[df["Genre"].str.contains(genre)]

    # Get the name of the highest rated movie
    print(f"Best movie in {genre} goes to:")
    best = contenders.loc[contenders["IMDB_Rating"] == contenders["IMDB_Rating"].max()].values[0]
    print(best)
    print()

# There may be a better way of doing this, but this is what I came up with given what I've been
# taught so far.
```

```python
# Next, find the director with the most movies
dir_counts = df["Director"].value_counts()
print(f"The director with the most movies is: {dir_counts.keys()[0]}")
```

```python
# Now we'll find the star with the most movies
stars = pd.concat([df[column].value_counts() for column in df if "Star" in column], axis = 1)
stars = stars.sum(axis=1)

print(f"The star that has appeared in the most movies is: {stars.keys()[0]}")
```

```python
df["Gross"].value_counts()
```

```python
# We can reuse the unique_genres variable from earlier

# Find the highest grossing movie in each genre
for genre in unique_genres:
    # Narrow it down to movies that are in this genre
    contenders = df[df["Genre"].str.contains(genre)]

    # Get the name of the highest grossing movie
    print(f"Highest grossing movie in {genre} goes to:")
    # Locate the entry with the highest grossing value, first formatting the gross string and then casting to float
    best = contenders.loc[contenders["Gross"].str.replace(',','').astype(float).idxmax()].values[0]
    print(best)
    print()
```

```python
directors = df["Director"].unique()

# Find the lowest grossing movie for each director
for director in directors:
    # Narrow it down to movies that are made by this director
    contenders = df[df["Director"].str.contains(director)]

    # Find the index of their worst grossing movies (any entries with NaN will be thrown out)
    # this gives a warning but in the interest of sanity we will ignore it
    min_idx = contenders["Gross"].str.replace(',','').astype(float).idxmin(skipna=True)
    if(np.isnan(min_idx)):
        continue

    # Get the name of the lowest grossing movie of the current director
    print(f"{director}'s lowest grossing movie is:")
    worst = contenders.loc[min_idx].values[0]
    print(worst)
    print()
```
