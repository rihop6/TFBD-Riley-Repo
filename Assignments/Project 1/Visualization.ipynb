# Step 4: Import Aggregated Data from SQL for Visualization

```python
# Import seaborn for visualization
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


stocks = pd.read_csv("stocks.csv")
cumulative_returns = pd.read_csv("cumulative_returns.csv")
```

```python
# Check returns data
cumulative_returns.head()
```

Plot the stocks dataframe information in bar plots

```python
# List of columns to plot
columns = ['total_return','volatility','avg_return','sharpe_ratio']

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(30, 6), sharey=True)

# Plot each bar plot
i = 0
for col in columns:
    sns.barplot(x='symbol', hue='symbol', y=col, data=stocks, ax=axs[i])
    axs[i].set_title(col)
    i += 1

# Adjust layout
plt.tight_layout()
plt.show()
```

Plot all cumulatives side by side

```python
wgmi = cumulative_returns.loc[cumulative_returns["Symbol"] == "WGMI"]
nvda = cumulative_returns.loc[cumulative_returns["Symbol"] == "NVDA"]
forex = cumulative_returns.loc[cumulative_returns["Symbol"] == "C:EURUSD"]
ndx = cumulative_returns.loc[cumulative_returns["Symbol"] == "NDX"]

# Make a figure and plot both lines
plt.figure(figsize=(10, 6))
fig, axs = plt.subplots(1, 4, figsize=(30, 6))
axs[0].plot(wgmi["Date"], wgmi["Cumulative Return"], label='WGMI', marker='x')
axs[0].set_title('WGMI')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Cumulative Return %')
axs[0].legend()

axs[1].plot(nvda["Date"], nvda["Cumulative Return"], label='NVDA', marker='x', c='orange')
axs[1].set_title('NVDA')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Cumulative Return %')
axs[1].legend()

axs[2].plot(forex["Date"], forex["Cumulative Return"], label='C:EURUSD', marker='x', c='green')
axs[2].set_title('C:EURUSD')
axs[2].set_xlabel('Date')
axs[2].set_ylabel('Cumulative Return %')
axs[2].legend()

axs[3].plot(ndx["Date"], ndx["Cumulative Return"], label='NDX', marker='x', c='teal')
axs[3].set_title('NDX')
axs[3].set_xlabel('Date')
axs[3].set_ylabel('Cumulative Return %')
axs[3].legend()

axs[0].set_ylim(0, 120)
axs[1].set_ylim(0, 225)
axs[2].set_ylim(-1, 1.2)
axs[3].set_ylim(-1, 20)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

Plot wgmi vs nvda

```python
# Make a figure and plot both lines
plt.figure(figsize=(10, 6))
plt.plot(wgmi["Date"], wgmi["Cumulative Return"], label='WGMI', marker='x')
plt.plot(nvda["Date"], nvda["Cumulative Return"], label='NVDA', marker='x')

# Add title and labels
plt.title('WGMI vs NVDA cumulative returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')

# Add legend
plt.legend()

# Show plot
plt.show()
```
